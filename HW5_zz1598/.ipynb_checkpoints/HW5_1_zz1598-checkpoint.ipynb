## The probability and cumulative distributions of riders' ages

1. Get the age's probability distribution(≈pdf), and cdf(normalization needed).
2. Use **scipy.stats.norm.cdf(x, loc=mean, scale=std)** to get y for the model normal distribution for comparison

```python
age_mean = df2.age.mean()
age_std = df2.age.std()

#cs=df2.age.groupby(pd.cut(df2.age, bins)).agg([count_nonzero]).cumsum()
age_distr = df2.age.groupby(df2.age).agg([count_nonzero])
age_prob_distr = age_distr / len(df2.age)  #normalize frequencies(pdf)
age_csum =  df2.age.groupby(df2.age).agg([count_nonzero]).cumsum()
age_csum_norm = age_csum / age_csum.max()   #normalize cumsum age(cdf)
```

```python
print ('mean age: ', age_mean)
print ('standard deviation: ', age_std)

print (len(df2.age))
print (age_csum.max())
```

# 2-1. Compare Sample Age Distribution with a Normal Distribution

```python
#model nomal distribution with mean=age_mean, std=age_std
x = list(set(list(df2.age)))
y = scipy.stats.norm.cdf(x, loc=age_mean, scale=age_std)

# calculate the absolute difference batween cdf of obervations and cdf of model normal distribution
d = np.abs(age_csum_norm.count_nonzero - y)
# np.abs(cs1/cs1.max() - cs2/cs2.max())  -- if for two sample comparison
```

```python
fig = pl.figure(figsize(12,12))
#fig, axes = pl.subplots(nrows=2, ncols=2)   #when plotting pic with same x and y labels?

ax1 = fig.add_subplot(221)  # 2rows, 2colums, 1st fig
ax1.plot(age_distr)
pl.title('Fig2a - Age Distribution of bikers')
pl.ylabel('Frequency of Age Counts')
pl.xlabel('Age')

ax2 = fig.add_subplot(222)
ax2.plot(age_prob_distr)
pl.title('Fig2b - Age Probability Distribution of bikers')
pl.ylabel('Probability')
pl.xlabel('Age')

ax3 = fig.add_subplot(223)
ax3.plot(x, age_csum_norm)
pl.title('Fig2c - Age Cumulative Distribution of bikers')
#pl.plot(bins1[:-1] + 5, age_csum_norm, label = ')   #using bins0 or bins1 or not using bins?
pl.ylabel('Probability')
pl.xlabel('Age')

ax4 = fig.add_subplot(224)
ax4.plot(d, 'r-')
pl.title('Fig2d - Absolute Difference of Observation from Model Normal Distributions')
pl.ylabel('Probability')
pl.xlabel('Age')
```

```python
# I want to compare the difference
fig = pl.figure(figsize(8,5))
ax5 = fig.add_subplot(111)
ax5.plot(x, age_csum_norm, label='Cumu Age Sample Distribution')
ax5.plot(x, y, label='Cumu Model Normal Distribution')
ax5.plot(x, d, label='Absolute Difference' )

pl.title('Fig3 - Difference between Cumulative Sample Distribution and the Model Normal Distribution')
pl.xlabel('Age- mu=age_mean, sigma=age_std')
pl.ylabel('Probability(Normalized Frequency of Counts)')
pl.legend(loc='right')
```

## 2-1a. KS tests to compare Sample Distribution to a Model Normal Distribution

http://docs.scipy.org/doc/scipy-0.15.1/reference/generated/scipy.stats.kstest.html


### $\alpha$ = 0.05
### Null Hypothesis
### H0: The sample age distribution comes from a normal distribution with mean 39.7119689 and standard deviation 11.7340092. 

```python
ks_norm = scipy.stats.kstest(df2.age.values, 'norm', args=(age_mean, age_std))
# for two sample: ks = scipy.stats.ks_2samp(df2.age1[~np.isnan(df2.age1)], df2.age2[~np.isnan(df2.age2)])

print (ks_norm)
```

### the KS test's p-value ≈ 0 < 0.05   ($\alpha$ = 0.05)
### So the null hypothesis is rejected. We conclude that the sample age distribution does not come from a normal distribution with mean 39.7119689 and standard deviation 11.7340092.

## 2-1b. Anderson's  test

### $\alpha$ = 0.05
### Null Hypothesis
### H0: The sample age distribution comes from a normal distribution with mean 39.7119689 and standard deviation 11.7340092. 

```python
# perform Anderson's Test
ad_norm = scipy.stats.anderson(df2.age, 'norm')   #two sided?
print (ad_norm)
```

### Interpreting the result of Anderson's Test:
### the sample  AD statistics ≈ 1838.280 **>** 0.787 that corresponds the our significant level $\alpha$=0.05
### So we reject the null hypothesis. We conclude that the sample age distribution does not come from a normal distribution with mean 39.7119689 and standard deviation 11.7340092.

# 2-2. Compare age distribution with an Exponential Distribution

```python
# y of model exponential distribution(cmf)  
y_exp = scipy.stats.expon.cdf(x, loc=16, scale=30)

# calculate the absolute difference batween cdf of obervations and cdf of model poisson distribution
d_exp = np.abs(age_csum_norm.count_nonzero - y_exp)
```

```python
# I want to compare the difference
fig_e = pl.figure(figsize(8,5))
ax_e = fig_e.add_subplot(111)
ax_e.plot(x, age_csum_norm, label='Cumu Age Sample Distribution')
ax_e.plot(x, y_exp, label='Cumu Model Exponential Distribution')
ax_e.plot(x, d_exp, label='Absolute Difference' )

pl.title('Fig3 - Difference between Cumulative Sample Distribution and the Model Expon Distribution')
pl.xlabel('Age- mu=age_mean, sigma=age_std')
pl.ylabel('Probability(Normalized Frequency of Counts)')
pl.legend(loc='right',labelspacing=0.3, prop={'size':9})
```

```python
# Perform Chi-Squared Test using pmf(probability mass function)
# each age as one category
y_pmf_poi = scipy.stats.poisson.pmf(x, age_mean)
# degree of freedom = #observations - parameter -1 = age_csum.max() - 1 -1
chiq_poi = scipy.stats.chisquare(f_obs=age_prob_distr, f_exp=y_pmf_poi, ddof=0, axis=0) 
print (chiq_poi)
```

```python
# Perform Chi-Squared Test using pmf(probability mass function) with resized bins1(17 categories)
age_distr_b1 = df2.age.groupby(pd.cut(df2.age, bins1)).agg([count_nonzero])
age_prob_distr_b1 = age_distr_bins1 / len(df2.age)

y_pmf_poi_b1 = scipy.stats.poisson.pmf(bins1, age_mean)
# degree of freedom = #observations - parameter -1 = age_csum.max() - 1 -1
chiq_poi_b1 = scipy.stats.chisquare(f_obs=age_prob_distr_b1, f_exp=y_pmf_poi_b1, ddof=1, axis=0) 
print (chiq_poi_bins1)
```

### Interpreting result of the Chisquared Test
