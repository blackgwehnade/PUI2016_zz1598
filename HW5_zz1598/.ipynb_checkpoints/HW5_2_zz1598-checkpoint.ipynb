## there is an analytical solution for the OLS (ordinary least square) regression which is coded below. please try to read the code and understand the formula! 


````python
allmales['hispanic'].loc[:,'Median income'][1]  #Locate cell: hispanic > Median Income > Total
````

````python
#this is the analytical best regression line. 
# it can be proved (ask me about it if you want!)
# run the cell below to get the beta0 and beta1 values

all_males=np.array([0., allmales['all'].loc[:,'Median income'][1],  #dataframe 'all', row0, column'Median i', row'value'
                    allmales['white'].loc[:,'Median income'][1],
                    allmales['black'].loc[:,'Median income'][1],
                     allmales['hispanic'].loc[:,'Median income'][1],
                    allmales['asian'].loc[:,'Median income'][1]])

all_females=np.array([0.] + [allfemales[k].loc[:,'Median income'][1] \
                             for k in ['all', 'white', 'black',
                                     'hispanic','asian']])
cor = stats.spearmanr(all_males, all_females)
beta1 = cor[0] * all_females.std() / all_males.std()
beta0 = all_females.mean() - beta1 * all_males.mean()
beta0, beta1
````

## Fit and Plot-- data points, 1:1 line, ols regression line, and analytical regression line

````python
# plot the data points for each income soure for each race
for race in race_lst:  
    y_f_all = np.array([allfemales[race]['Median income'][:].values]) #header and index column doesn't count
    x_m_all = np.array([allmales[race]['Median income'][:].values]) #dataframe 'all', column'Median i', row'value'
    mask_fm0 = ~((isnan(y_f_all)) + (isnan(x_m_all)))
    y_f_all = y_f_all[mask_fm0]
    x_m_all = x_m_all[mask_fm0]
    
    pl.plot(x_m_all, y_f_all, '.', ms=10, label=race)

# fit a 1:1 line
pl.plot((0,45000),(0,45000), 'b')
    
pl.title('Ratio of Midian Income(Female vs Male by Race) ')
pl.xlabel("Male Midian Income from all Sources")
pl.ylabel("Female Midian Income from all Sources")
pl.legend(loc='best', labelspacing=0.3, prop={'size':9})
````

````python
#all_males_allincomes = np.array(
#    np.concatenate([allmales['all']['Median income'][:].values,
#                            allmales['white']['Median income'][:].values,
#                           allmales['black']['Median income'][:].values,
#                           allmales['hispanic']['Median income'][:].values,
#                       allmales['asian']['Median income'][:].values]).flatten())

# loop version of the above
all_males_allincomes = np.array(
    np.concatenate([allmales[r]['Median income'][:].values for r in allmales.keys()]).flatten())

all_females_allincomes = np.array(
    np.concatenate([allfemales[r]['Median income'][:].values for r in allfemales.keys()]).flatten())

#the mask is False where either male or female value is NaN. 
#Notice that True is really just >1, and False really 0. 
#Adding up: True if both arrays are not Nan!

mask_fm = ~((np.isnan(all_males_allincomes)) + (np.isnan(all_females_allincomes)))
# now you can use mask as an index (broadcasting) to select the non NaN values
````

````python
X_m_all = all_males_allincomes[mask_fm]
Y_f_all = all_females_allincomes[mask_fm]

df_all = pd.DataFrame({'all_mid_female':Y_f_all, 'all_mid_male':X_m_all})

lm_all = smf.ols(formula = 'Y_f_all ~ X_m_all', data=df_all).fit()
df_all.loc[:,"error"]=lm_all.resid
df_all.loc[:,"y_f_fit"]=lm_all.predict()
df_all

df_all.plot(kind='scatter',x='all_mid_male',y='all_mid_female')
pl.plot(df_all.all_mid_male,lm_all.predict(),'r', linewidth=1.0)

pl.title('Ratio of Midian Income with fitted line(F vs M by Race) ')
````

````python
# Analytical Solution
all_males_all = np.array([0.] + X_m_all)
all_females_all = np.array([0.] + Y_f_all)

cor_all = stats.spearmanr(all_males_all, all_females_all)
beta1_all = cor[0] * all_females_all.std() / all_males_all.std()
beta0_all = all_females_all.mean() - beta1_all * all_males_all.mean()
beta0_all, beta1_all
````

### Observation:
### 1. Line comparisons
* The **1:1 lines** fit similar to the one with only the TOTAL median income-- almost all the data points fall below the line. This follows the predicted situation that a woman is likely to earn less given a man's earning.
* The **OLS regression line** fits better when we include all the income sources-- when using only the total median income, the line does not intersect any data point at all.
* The **analytical regression line** fits better when data points of different income sources are included in the model.
* With all forms of income, the gap between the OLS and the Analytical regression lines decreases(almost overlapped), while with only the Total Income, the two fitted lines intersect each other. 

### 2. High legverage points
* According to the captain graph, almost all the data points for black people are above both the ols line and the analytical line. This means that given a black male's median income, these two methods would underestimate the earning of a black female. Similarly, these two methods are likely to overestimate the earning of a white female given the earning of a white male.
* There are several outliars that are the extreme cases of the underestimates and overestimates mentioned above.

## Now pick a stipend (maybe your ideal stipend).
## and predict what a woman in a position where a man recieves your ideal stipend would make



````python
lm_all.params
````

````python
def ols_reg(x):
    y = lm_all.params[0] + lm_all.params[1]*x
    return y

def ana_reg(x):
    y = beta0_all + beta1_all*x
    return y
````

````python
# I would use the average of the ols and the analytical regression models
stipend =  100000
earn_f_ols = ols_reg(stipend)
earn_f_ana = ana_reg(stipend)
earn_f = (earn_f_ols + earn_f_ana)/ 2

print ('ols estimated female\'s earning: ', earn_f_ols)
print ('analytical estimated female\'s earning: ', earn_f_ana)
print ('estimated female\'s earning: ', earn_f)
````
